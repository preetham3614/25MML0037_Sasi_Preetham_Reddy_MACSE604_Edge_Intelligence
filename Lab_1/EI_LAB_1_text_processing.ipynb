```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.gutenberg.org/cache/epub/84/pg84-images.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

Here in the above we can see that the data is \<p> tags . so we scrape the data that is present in between the \<p> \</p> tags. here for this use a method call find_all which will generate all the texts that is present in the paragaph tags

```python
quotes = []
quote_boxes = soup.find_all('p')
for i , box in enumerate(quote_boxes):
    text = box.get_text(strip=True)
    if text:
        quotes.append(text)

for q in quotes[:5]:
    print(q)
```

# **Step 2**

checking the number of paragraphs are scrapped and check if there are any duplicate paragraphs

```python
# final quotes the came fter scraping
len(quotes)
```

```python
# removing the duplicate quotes using the set dataset

text_data = list(set(quotes))

print(len(text_data))
```

```python
# checking sample of 5 sentences
text_data[:5]
```

Observation : As we can see from the above 5 sample we can observer that there are many irrelevent charecters so we remove the characters like  
  \n  ,  \r  ,  -  ,  !  ,  ;  ,  “  ,  ”  
  

# **Step 3**

As per the observation above we replace the irrelevent characters

```python
# combine all the text into a single variable

total_text = " ".join(text_data)
```

```python
# repace characters :  \n , \r , - , ! , ; , “ , ” with ""

total_text = total_text.replace("\n" , "")
```

```python
total_text = total_text.replace("\r" , "")
```

```python
total_text = total_text.replace("!" , "")
total_text = total_text.replace("“" , "")
total_text = total_text.replace("”" , "")
```

```python
total_text = total_text.replace("-","")
total_text = total_text.replace(";","")
total_text = total_text.replace(",","")
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from nltk.stem import PorterStemmer , WordNetLemmatizer
from nltk import pos_tag
```

# **Step 4 **

Tokenize the String


```python
# tokenize the words . this will split the paragraph into words which are also called as tokens
words = word_tokenize(total_text)
```

```python
# error occured due to insufficient libraries

nltk.download("punkt_tab")
words = word_tokenize(total_text)
```

**Step 5**

Additional Pre Processing on the words inorder to make words even more better for parts of speech tagging



1.   Remove Stop words
2.   Apply Stemming
3.   Apply Lemmatization



```python
# remove the stop words . here we are filtering the stopwords like is , that ,  are , etc..
nltk.download("stopwords")
filtered_words = [word for word in words if word not in stopwords.words('english')]
```

```python
# apply stemmer . this process will trim trailing part of the word . ex : if we have running , runs , ran , that will converted into run

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words ]
```

```python
# apply lemmatizer . This will convert the word to its root word . if we have better ,best these will be converted to good
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word)  for word  in stemmed_words]
```

# **Step 6**

Apply POS Tagging and print the tags

```python
# apply pos tagging. Pos Tagging is nothing but identifying the parts of speech of the given word
nltk.download("averaged_perceptron_tagger_eng")
tags = pos_tag(lemmatized_words)
```
