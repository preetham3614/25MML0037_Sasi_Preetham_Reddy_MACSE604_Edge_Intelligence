```python
import cv2
#Open default camera(0 = laptop webcam)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read() #read frame from camera

    if not ret:
        print("Failed to grab frame")
        break

    cv2.imshow("laptop camera",frame) # show the frame

    #press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
import cv2

cap = cv2.VideoCapture("video_1.mp4")

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Video",frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Display Video Properties

import cv2

cap = cv2.VideoCapture("video_1.mp4")

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

print("Total frames:",total_frames)
print("fps:",fps)
print("width",width)
print("height:",height)

cap.release()
```

```python
#Save 100th frame

import cv2

cap = cv2.VideoCapture("video_1.mp4")

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1

    # Processing every 100th frame
    if frame_count % 100 == 0:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow("frame",gray)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.rectangle(frame, (100,100),(300,300),(0,255,0),3)
    cv2.circle(frame,(200,200),50,(255,0,0),3)

    cv2.putText(frame,"OpenCV practice",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)

    cv2.imshow("Shapes",frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    edges = cv2.Canny(frame,100,200)

    cv2.imshow("Edges",edges)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    negative = cv2.bitwise_not(frame)

    cv2.imshow("Edges",negative)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Detecting movement
import cv2

cap = cv2.VideoCapture(0)

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while True:
    diff = cv2.absdiff(frame1,frame2)
    gray = cv2.cvtColor(diff,cv2.COLOR_BGR2GRAY)

    blur = cv2.GaussianBlur(gray,(5,5),0)
    _,thresh = cv2.threshold(blur,20,255,cv2.THRESH_BINARY)

    cv2.imshow("Motion",thresh)

    frame1 = frame2
    ret,frame2 = cap.read()
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
